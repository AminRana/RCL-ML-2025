```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the csv file in df
df = pd.read_csv('student_scores.csv');
```

```python
# Checking for any missing values in the dataset
print("DB Description: \n",df.describe())
print("Missing Values: \n",df.isnull().sum())
```

```python
# 1. Line graph: Study Hours vs Test Score
plt.plot(df['Study Hours'], df['Test Score'], marker='o', color='blue')
plt.title('Test Score vs Study Hours')
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.grid(True)
plt.show()
```

```python
# 2. Bar chart: Each Student's Test Score
plt.bar(df['Student'], df['Test Score'], color='green')
plt.title("Students' Test Scores")
plt.xlabel('Student')
plt.ylabel('Test Score')
plt.xticks(df['Student'])
plt.show()
```

```python
# 3. Histogram: Distribution of Test Scores
plt.hist(df['Test Score'], bins=5, color='purple', edgecolor='black')
plt.title('Histogram of Test Scores')
plt.xlabel('Test Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Histogram: Distribution of Study Hours
plt.hist(df['Study Hours'], bins=5, color='yellow', edgecolor='black')
plt.title('Histogram of Study Hours')
plt.xlabel('Study Hours')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Scatter plot: Study Hours vs Test Score
plt.scatter(df['Study Hours'], df['Test Score'], color='red', marker='o')
plt.title('Scatter Plot: Study Hours vs Test Score')
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.grid(True)
plt.show()
```

```python
# Seaborn lineplot with smoothing (LOWESS)
sns.lmplot(x='Study Hours', y='Test Score', data=df, lowess=True, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
plt.title('Smoothed Line: Study Hours vs Test Score (LOWESS)')
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.grid(True)
plt.show()
```
