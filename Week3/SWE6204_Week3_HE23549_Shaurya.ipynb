```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
# Preparing Data
# Numerical Data (Discrete and Continuous)
numerical_data = {
"Heights": [165, 170, 168, 172, 177], # Continuous
"Purchases": [2, 3, 1, 5, 4] # Discrete
}

# Categorical Data
categorical_data = {
"Gender": ["Male", "Female", "Female", "Male", "Male"],
"State": ["TX", "CA", "NY", "CA", "TX"]
}

# Ordinal Data
ordinal_data = {
"Movie Rating": [5, 3, 4, 2, 1] # 1 to 5 scale
}
```

```python
# Display data types
print("Numerical Data:", numerical_data)
print("Categorical Data:", categorical_data)
print("Ordinal Data:", ordinal_data)
```

```python
data = [0, 2, 3, 2, 1, 0, 0, 2, 0]

# Calculate Mean
mean_value = np.mean(data)
print("Mean:", mean_value)

# Calculate Median
median_value = np.median(data)
print("Median:", median_value)

# Calculate Mode
mode_value = stats.mode(data)
print("Mode:", mode_value)

# Calculate Variance
variance_value = np.var(data)
print("Variance:", variance_value)

# Calculate Standard Deviation
std_dev_value = np.std(data)
print("Standard Deviation:", std_dev_value)
```

```python
# Histogram for the dataset
plt.hist(data, bins=5, edgecolor='black')
plt.title("Histogram of Dataset")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Loading the csv file in df
df = pd.read_csv('demo_data.csv');
```

```python
# Checking for any missing values in the dataset
print("DB Description: \n",df.describe())
print("Missing Values: \n",df.isnull().sum())
```

```python
# Histogram of salaries
plt.hist(df['Salary'],bins=10,color='green',edgecolor='k')
plt.title('Histogram of Salaries')
plt.xlabel('Salary')
plt.ylabel('No. of Employees')
plt.show()
```

```python
# Histogram of salaries - Detailed
plt.hist(df['Salary'],bins=10,color='w',edgecolor='k')
plt.axvline(df['Salary'].mean(),color='g',linestyle='dotted',linewidth=5)
plt.axvline(df['Salary'].median(),color='b',linestyle='dotted',linewidth=5)
plt.axvline(df['Salary'].mode().values[0],color='y',linestyle='dotted',linewidth=5)
plt.title('Frequency Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

```python
plt.boxplot(df['Salary'])
plt.show()
```

```python
# to create subplots for showing multiple plots together by specifying the number of rows
f,(ax_box,ax_hist)=plt.subplots(2,sharex=True,gridspec_kw={"height_ratios":(0.5,2)})
# creating a boxplot by passig the data and giving the x values and stating the axis
sns.boxplot(data=df,x=df['Salary'],ax=ax_box)
ax_box.axvline(df['Salary'].mean(),color='k',linestyle='dotted',linewidth=2)
ax_box.axvline(df['Salary'].median(),color='g',linestyle='dotted',linewidth=2)
ax_box.axvline(df['Salary'].mode().values[0],color='y',linestyle='dotted',linewidth=2)
sns.histplot(data=df,x=df['Salary'],ax=ax_hist,kde=True)
ax_hist.axvline(df['Salary'].mean(),color='k',linestyle='dotted',linewidth=2)
ax_hist.axvline(df['Salary'].median(),color='g',linestyle='dotted',linewidth=2)
ax_hist.axvline(df['Salary'].mode().values[0],color='y',linestyle='dotted',linewidth=2)
plt.show()
```
